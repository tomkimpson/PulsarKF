## PulsarJ1713+0747
This notebook looks PSR J1713+0747 as a representative example of a PTA pulsar.


[PSR J1713+0747](https://www.atnf.csiro.au/research/pulsar/psrcat/proc_form.php?version=1.69&JName=JName&RaJ=RaJ&DecJ=DecJ&P0=P0&P1=P1&F0=F0&F1=F1&F2=F2&F3=F3&Dist=Dist&startUserDefined=true&c1_val=&c2_val=&c3_val=&c4_val=&sort_attr=jname&sort_order=asc&condition=&pulsar_names=J1713%2B0747&ephemeris=short&coords_unit=raj%2Fdecj&radius=&coords_1=&coords_2=&style=Long+with+last+digit+error&no_value=*&fsize=3&x_axis=&x_scale=linear&y_axis=&y_scale=linear&state=query&table_bottom.x=71&table_bottom.y=7) is one of 47 MSPs that make up the  NANOGrav dataset. See e.g. Fig 1 from [Alam et al. 2020](https://arxiv.org/pdf/2005.06495.pdf)


The two key parameters here are:

* Spin frequency: 218 Hz

* Spin frequency derivative: $-4 \times 10^{-16}$ s $^{-2}$


If we assume that the derivative of the spin frequency is ~constant, what does the frequency evolution look like over 10 years?




```python

import numpy as np

f0 = 218
fdot = -4e-16


Tend = 10 #years
year = 365*24*3600
t = np.linspace(0,10*year,int(1e3))

f = f0+ fdot*t #ie. Euler
print("dt = ", (t[1] - t[0])/year * 365)



import matplotlib.pyplot as plt 


plt.figure()

plt.plot(t/year,f/1e9)
noise = np.random.normal(0,3e-9,len(f))
fnoisy = f + noise
plt.plot(t/year,fnoisy/1e9)
plt.xlabel('T [years]')
plt.ylabel('f [nHz]')

print("The change in the frequency over {Tend} years is:", (max(f) - min(f))*1e9, " ns")
```
