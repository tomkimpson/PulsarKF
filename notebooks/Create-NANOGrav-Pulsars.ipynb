## Create PTA pulsars
This notebook gets a selection of pulsars and their parameters that are used in the NANOGrav PTA

### 1. Fetch all the pulsars 


First lets get all the pulsars (>3000) from the ANTF catalogue, along with a selection of the relevant parameters

```python
from psrqpy import QueryATNF
query = QueryATNF(params=['Name','PSRJ', 'F0','F1','RAJD','DecJD','Dist'])
df = query.pandas
```

### 2. Select only NANOGrav pulsars 

We are only really interested in those pulsars which are part of a PTA.

We will work with the 47 pulsars that make up the [NANOGrav PTA](https://arxiv.org/pdf/2005.06495.pdf) 

```python
NANOGrav_pulsars = ['J0023+0923',
                    'J0030+0451',
                    'J0340+4130',
                    'J0613-0200',
                    'J0636+5128',
                    'J0645+5158',
                    'J0740+6620',
                    'J0931-1902',
                    'J1012+5307',
                    'J1024-0719',
                    'J1125+7819',
                    'J1453+1902',
                    'J1455-3330',
                    'J1600-3053',
                    'J1614-2230',
                    'J1640+2224',
                    'J1643-1224',
                    'J1713+0747',
                    'J1738+0333',
                    'J1741+1351',
                    'J1744-1134',
                    'J1747-4036',
                    'J1832-0836',
                    'J1853+1303',
                    'J1857+0943', # renamed from B1855+09
                    'J1903+0327',
                    'J1909-3744',
                    'J1910+1256',
                    'J1911+1347',
                    'J1918-0642',
                    'J1923+2515',
                    'J1939+2134', # renamed from B1937+21
                    'J1944+0907',
                    'J1946+3417',
                    'J1955+2908', # renamed frm B1953+29
                    'J2010-1323',
                    'J2017+0603',
                    'J2033+1734',
                    'J2043+1711',
                    'J2145-0750',
                    'J2214+3000',
                    'J2229+2643',
                    'J2234+0611',
                    'J2234+0944',
                    'J2302+4442',
                    'J2317+1439',
                    'J2322+2057']

```

```python
df_NANOGrav = df.query("PSRJ in @NANOGrav_pulsars")
display(df_NANOGrav)
```

### 3. Plot the selected NANOGrav pulsars 

Lets plot the distribution of these pulsars on the sky

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np 




# Read RA, DEC data from file.
ra = df_NANOGrav["RAJD"].to_numpy()
dec = df_NANOGrav["DECJD"].to_numpy()

# Convert RA/DEC from decimal degrees to radians.
ra,dec = np.radians(ra), np.radians(dec)

# Make plot.
fig = plt.figure(figsize=(20, 20))
gs = gridspec.GridSpec(4, 2)
# Position plot in figure using gridspec.
ax = plt.subplot(gs[0], polar=True)
# Plot points.
ax.scatter(ra, dec, marker='o', c='C0', s=50, lw=0.) #ra is how far round, dec is how far out!
```

### 4. Clean up the dataframe, and save it to disk


Lets now do the following:

* Drop unnecessary columns
* Determine the $\gamma$ factor assuming n = 3
* Make all angles radians

```python
df_clean = df_NANOGrav[["PSRJ","F0","F1", "RAJD", "DECJD", "DIST"]].copy()
df_clean["RAJD"] = np.radians(df_clean["RAJD"])
df_clean["DECJD"] = np.radians(df_clean["DECJD"])

n = 3
df_clean["gamma"] = -df_clean["F1"]/df_clean["F0"]**n 
df_clean["n"] = n
```

```python
assert (df_clean["gamma"] > 0).all()  #all gammas should be negative
```

```python
df_clean.to_pickle('../data/NANOGrav_pulsars')
```
