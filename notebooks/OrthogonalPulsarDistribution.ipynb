## Generating points orthogonal to a direction

This notebook looks at how to generate $N$ points perpendicular to a propagation direction


```python
import numpy as np




source_theta = np.random.uniform(low=0,high=np.pi/2,size=1)
source_phi = np.random.uniform(low=0,high=2*np.pi,size=1)



def cartesian(theta,phi):


    x= np.sin(source_theta)*np.cos(source_phi)
    y = np.sin(source_theta)*np.sin(source_phi)
    z = np.cos(source_theta)

    return np.array([x,y,z])

GW_direction = cartesian(source_theta,source_phi)
    

```

```python
GW_direction = np.array([ 0.59500984,  0.09655469, -0.79789754])
```

```python

N = 10


pulsars = np.zeros((N,3))

for i in range(N):
    x = np.random.randn(3)  # take a random vector
    x -= x.dot(GW_direction) * GW_direction       # make it orthogonal to k
    x /= np.linalg.norm(x)  # normalize it

    pulsars[i,:] = x 


```

```python
y = np.cross(GW_direction, x)
```
